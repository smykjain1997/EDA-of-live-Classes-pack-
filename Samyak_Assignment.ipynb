```python
grade1=[]
for i in data["Batch"]:
    c=i.split("-")
    d=c[1].split(" [")
    grade1.append(d[0])
```

```python
g=[]
n=[]
for i in grade1:
    e=i.split(" ")
    #print(len(e))
    d=e[1].rstrip("s")
    #print(d)
    f=e[2]
    #print(d,f)
    if len(e)>3:
        k=e[2]+e[3]
        n=(d+" "+k)
        g.append(n)
    else:
        n=(d+" "+f)
        g.append(n)
```
